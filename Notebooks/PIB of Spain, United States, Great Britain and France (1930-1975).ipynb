```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Comparaciones_Internacionales_.csv', sep=";", skiprows=653, nrows=46)

df.rename(columns={'1929': 'Years'}, inplace=True)
df.rename(columns={'6562': 'Spain'}, inplace=True)
df.rename(columns={'10909': 'France'}, inplace=True)
df.rename(columns={'9866': 'UK'}, inplace=True)
df.rename(columns={'13522': 'EEUU'}, inplace=True)

df
```

```python
x = df.Years
y = df.Spain
y2 = df.France
y3 = df.UK
y4 = df.EEUU

print(x.shape,y.shape)
```

```python
# Gráfico
fig, ax = plt.subplots(figsize=(35,15), facecolor='white')
ax.set_facecolor('white')

# Grid
plt.grid(alpha=1, linewidth=0.4, color='grey', zorder=0)

# Lineas
ax.plot(x, y, label='España', color='red', linewidth=5, alpha= 1, zorder=5)
ax.plot(x, y2, label='France', color='blue', linewidth=5, alpha= 1, zorder=3)
ax.plot(x, y3, label='UK', color='green', linewidth=5, alpha= 1, zorder=3)
ax.plot(x, y4, label='EEUU', color='orange', linewidth=5, alpha= 1, zorder=3)

# Ejes
plt.yticks(np.arange(0, 40000, 5000))
ax.set_xlim(left=1930)
ax.set_xlim(right=1975)
plt.tick_params(labelsize=25)

# Label
ax.tick_params(colors='black', length=15, width=1.5, which='both')

# Bordes
for axis in ['bottom', 'top', 'left', 'right']:
  ax.spines[axis].set_color('black')
  ax.spines[axis].set_linewidth(3)
  ax.spines[axis].set_zorder(5)

# Texto
fig.text(0.513, 0.03, 'Años', ha='center', va='center', fontsize=30, color='black', weight='bold')
fig.text(0.07, 0.5, 'Dólares GEKS de 2021', rotation=90, ha='center', va='center', fontsize=30, color='black', weight='bold')

# Leyenda
legend = ax.legend(['España', 'Francia', 'Reino Unido', 'Estados Unidos'], borderaxespad=1, loc='upper left', fontsize=25, facecolor='white', edgecolor='black', borderpad=0.8)
for text in legend.get_texts():
    text.set_fontweight("normal")
    text.set_color("black")
    text.set_fontsize(25)
    
fig.savefig("Figura 1", dpi=200, bbox_inches='tight', pad_inches=1)

plt.show()
```
