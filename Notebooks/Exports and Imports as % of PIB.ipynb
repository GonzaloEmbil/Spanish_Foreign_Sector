```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Importaciones-Exportaciones.csv', sep=";")

df
```

```python
x = df.Years
y = df.Exp
y2 = df.Imp

print(x.shape,y.shape, y2.shape)
```

```python
# Gráfico
fig, ax = plt.subplots(figsize=(35,15), facecolor='white')
ax.set_facecolor('white')

# Ejes
ax.set_ylim(top=35)
ax.set_ylim(bottom=0)
plt.xticks([0, 5, 10, 15, 20, 25, 30, 35])
ax.set_xlim(left=1850)
ax.set_xlim(right=2015)
plt.xticks([1850, 1860, 1870, 1880, 1890, 1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010])
plt.tick_params(labelsize=25)

# Grid
plt.grid(alpha=1, linewidth=0.4, color='grey', zorder=0)

# Lineas
ax.plot(x, y, label='Exp', color='green', linewidth=5, alpha= 1, zorder=3)
ax.plot(x, y2, label='Imp', color='red', linewidth=5, alpha= 1, zorder=3)

# Sombreado
plt.fill_between(x, y, y2, where=(y2 <= y), facecolor="green", alpha= 0.2, zorder=2)
plt.fill_between(x, y, y2, where=(y < y2), facecolor="red", alpha= 0.2, zorder=2)
plt.axvspan(1939, 1959, color='grey', alpha=0.25, lw=0, zorder=1)
plt.axvspan(1959, 1975, color='#e6d690', alpha=0.25, lw=0, zorder=1)

# Label
ax.tick_params(colors='black', length=15, width=1.5, which='both')

# Bordes
for axis in ['bottom', 'top', 'left', 'right']:
  ax.spines[axis].set_color('black')
  ax.spines[axis].set_linewidth(3)
  ax.spines[axis].set_zorder(5)

# Texto
fig.text(0.513, 0.03, 'Años', ha='center', va='center', fontsize=30, color='black', weight='bold')
fig.text(0.09, 0.5, '% del PIB', rotation=90, ha='center', va='center', fontsize=30, color='black', weight='bold')
fig.text(0.59, 0.7, 'Autarquía', rotation=90, ha='center', va='center', fontsize=30, color='black')
fig.text(0.67, 0.7, 'Desarrollismo', rotation=90, ha='center', va='center', fontsize=30, color='black')

# Leyenda
legend = ax.legend(['Exportaciones (% del PIB)', 'Importaciones (% del PIB)'], borderaxespad=1, loc='upper left', fontsize=25, facecolor='white', edgecolor='black', borderpad=0.8)
for text in legend.get_texts():
    text.set_fontweight("normal")
    text.set_color("black")
    text.set_fontsize(25)
    
fig.savefig("Figura 6", dpi=200, bbox_inches='tight', pad_inches=1)

plt.show()
```
