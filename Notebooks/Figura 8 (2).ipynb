```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1940, 1959, 5)

serie1 = [26.01, 29.19, 18.84, 10.76]
serie2 = [0.98, 0.97, 0.49, 0.40]
serie3 = [3.05, 2.02, 1.83, 2.03]
serie4 = [20.58, 24.54, 20.05, 18.14]
serie5 = [13.38, 11.54, 12.07, 16.13]
serie6 = [20.51, 14.56, 23.86, 19.64]
serie7 = [3.53, 4.83, 7.63, 9.73]
serie8 = [8.4, 8.89, 14.65, 17.96]
serie9 = [3.57, 3.47, 4.58, 5.22]

y = np.vstack([serie1, serie2, serie3, serie4, serie5, serie6, serie7, serie8, serie9])

cols = ['#fff0f5', '#d1a4a4', '#8a0f0f', 'lightblue', '#4682b4', '#416787', 'yellowgreen', '#308446', '#0b6730']

fig, ax = plt.subplots(figsize=(40,15), facecolor='white')
ax.set_facecolor('white')

ax.stackplot(x, y, labels = ["Alimentos, bebidas y tabaco", "Textiles", "Otros bienes de consumo manufacturados", "Materias primas", "Productos químicos", "Combustibles", "Metales y sus manufacturas", "Maquinaria y aparatos", "Elementos de transporte"], colors = cols, alpha = 0.9)

# Ejes
ax.set_ylim(top=100)
ax.set(xlim = (min(x), max(x)), xticks = x)
plt.tick_params(labelsize=25)

# Grid
plt.grid(alpha=0, linewidth=0.4, color='grey', zorder=0)

# Label
ax.tick_params(labelsize=40, colors='black', length=15, width=1.5, which='both')
plt.xticks([1940, 1945, 1950, 1955], ['1940-1944', '1945-1949', '1950-1954', '1955-1959'])

# Bordes
for axis in ['bottom', 'top', 'left', 'right']:
  ax.spines[axis].set_color('black')
  ax.spines[axis].set_linewidth(3)
  ax.spines[axis].set_zorder(5)

# Texto
fig.text(0.513, 0.023, 'Años', ha='center', va='center', fontsize=45, color='black', weight='bold')
fig.text(0.08, 0.5, '% de las importaciones', rotation=90, ha='center', va='center', fontsize=45, color='black', weight='bold')

# Leyenda
legend = ax.legend(bbox_to_anchor=(1, 0.8), loc='upper left', borderaxespad=1, fontsize=30, facecolor='white', edgecolor='black', borderpad=0.8)
for text in legend.get_texts():
    text.set_fontweight("normal")
    text.set_color("black")
    text.set_fontsize(30)

fig.savefig("Figura 8", dpi=200, bbox_inches='tight', pad_inches=1)

plt.show()
```
